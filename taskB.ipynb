```python
# plot modules
%matplotlib inline
from matplotlib import pyplot as plt
# distributions
from scipy import stats as stats
```

```python
S = 10**5
N = 5
lam = 0.25
```

```python
g = stats.gamma.rvs(size=S, a=N, loc=0, scale=lam)
```

```python
plt.hist(g)
plt.show()
```

```python
es = [sum(stats.expon.rvs(size=N, loc=0, scale=lam)) for s in range(S)]
```

```python
plt.hist(es)
plt.show()
```
